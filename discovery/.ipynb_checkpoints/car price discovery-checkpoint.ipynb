```python
df['tvalues']  = df['coef']/df['std err']
df.round(4)
```

#### Проводим T-тест, чтобы определить, какие коэффициенты значимы, а какие нет

$\alpha$ = 0.05

$n$ = 205

$\kappa$ = 5 (+1)


```python
t_critical = stats.t.ppf(1-0.05/2, 205 - 6 )
t_critical.round(3)
```

```python
df['значимости'] = df.apply(lambda x : 'Значимо' if np.abs(x['tvalues']) > t_critical else 'НЕ значимо', axis = 1 )
df
```

$\mathbb{All}$ $\mathbb{coefficients}$ $\mathbb{are}$ $\mathbb{significant}$ $\mathbb{except}$ $\mathbb{enginesize}$ $\mathbb{and}$ $\mathbb{doornumber}$

### Проверим значимость влияния категориальной переменной carbody на регрессию

$$
H_0: \beta_(hardtop) = \beta_(hatchback) = \beta_(sedan) = \beta_(wagon) = \beta_(convertible) = 0
$$

```python
unre_model = smf.ols(data = df_all, formula = 'np.log(price)~carbody_hardtop+ carbody_hatchback+carbody_sedan+carbody_wagon+carbody_convertible + fueltype + enginesize + doornumber+ horsepower + wheelbase').fit()
re_model = smf.ols(data = df_all, formula = 'np.log(price)~ fueltype + enginesize + doornumber+ horsepower + wheelbase').fit()
```

```python
print(unre_model.f_test("carbody_hardtop=carbody_hatchback=carbody_sedan=carbody_wagon=carbody_convertible=0"))
```

```python
n=len(df_all) 
DF=unre_model.df_resid 
q =5 
R2_ur=unre_model.rsquared 
R2_r = re_model.rsquared
```

```python
F_nabl = ((R2_ur - R2_r)/(1-R2_ur))*DF/q
F_nabl
```

```python
f_test = unre_model.f_test("carbody_hardtop=carbody_hatchback=carbody_sedan=carbody_wagon=carbody_convertible=0")
```

```python
f_test.fvalue
```

```python
alpha = 0.05
F_cr = stats.f.ppf(1-0.05,q,DF)
```

```python
if (F_nabl > F_cr ):
    print('Отвергаем H0, коэффициенты совместно значимы')
else:
    print('Не отвергаем H0, коэффициенты совместно незначимы')
```

## F-тест

$$
H_0: carbody = fueltype = enginesize = doornumber = horsepower = wheelbase = 0
$$

$$
H_1: carbody^2 + fueltype^2 + enginesize^2 + doornumber^2 + horsepower^2 + wheelbase^2 > 0
$$

```python
F= unre_model.fvalue
print(F)
```

```python
F_cr = f.ppf(q=1-0.05, dfn=unre_model.df_model, dfd=unre_model.df_resid)
print(F_cr)
```

```python
if (F > F_cr ):
    print('Отвергаем H0, коэффициенты совместно значимы')
else:
    print('Не отвергаем H0, коэффициенты совместно незначимы')
```

$\clubsuit$ Составленная модель регресси в целом явлется статистически значимой

## Проверка на мультиколлениарность 

```python
VIFS = pd.DataFrame({'VIF': [variance_inflation_factor(car_price.model.exog, i) for i in range(int(car_price.df_model+1))]}, index=car_price.model.exog_names)
VIFS.drop(index='Intercept', inplace=True)
VIFS
```

$\diamond$ VIF больше 1, что свидетельствует о наличии мультиколленеарности, значит существует значимая зависимость между перемнными, наши выводы могут быть не совсем корректны, поэтому в последующих исследованиях надо будет добавлять новые переменные или проводить более тщательные тесты. Из исходных данных можно сделать вывод, что количество лошадиных сил влияет на цены седанов, так как седаны разгоняются быстрей из-за своей обтекаемой формы (меньше сопротивление воздуха)

## Диагностический тест на гетероскедастичность

$$
e_i^2 = \beta_0 + \beta_1(carbody) + \beta_2(fueltype) + \beta_3(enginesize) + \beta_4(doornumber) + \beta_5(horsepower) + \beta_6(wheelbase)
$$

$$
H_0: \beta_0 = \beta_1(carbody) = \beta_2(fueltype) = \beta_3(enginesize) = \beta_4(doornumber) = \beta_5(horsepower) = \beta_6(wheelbase) = 0
$$

```python
df['res2']=unre_model.resid**2
df['res']=unre_model.resid
```

```python
ei_model= smf.ols(data=df, formula='res2~carbody + fueltype + enginesize + doornumber+ horsepower + wheelbase').fit()
```

```python
St = np.round(len(df)*ei_model.rsquared,2)
St
```

```python
Hi2 = stats.chi2.ppf(1-0.05,6)
Hi2.round(2)
```

 
$\heartsuit$ ST>Hi2 $\rightarrow$ H_0 отвергаем. 

$\heartsuit$ Тест на гетероскедастичность положительный

$\blacksquare$ Беря во внимание, что проведенный тест указывает на гетероскедастичность, в нашей модели оценки стандартных ошибок могут быть либо недооценены, либо переоценены. Такая ситуация может привести к неверной интерпетации гипотез, а также повышают риск ошибки в построении новых гипотез относительно полученных данных. Для более полного исследования данной проблемы, проведем еще один тест

## Тест на серийную корреляцию

$$
carprice_t = \beta_0 + \beta_1(carbody_t) + \beta_2(fueltype_t) + \beta_3(enginesize_t) + \beta_4(doornumber_t) + \beta_5(horsepower_t) + \beta_6(wheelbase_t) + u_t
$$

#### $\blacktriangle$ Предположим, что

$$
u_t = u_0 + \rho*u_(t-1)
$$

$$
H_0: \rho = 0
$$

$$
H_1: \rho \neq 0
$$

```python
resSq = sum(unre_model.resid**2)
```

```python
sum_resid = 0
for i in range(len(unre_model.resid)-1):
    sum_resid+=(unre_model.resid[i+1]-unre_model.resid[i])**2
```

```python
sum_resid/resSq
```

$\heartsuit$ Гипотеза H_0 отвергается, есть серийная корреляция. Так как мы отвергли Н_0, потому что у нас значение теста получилось 1.0011, получается, что в остатках нашей модели линейной регрессии присутствует автокорреляция. Важно отметить, что она не очень сильная.
