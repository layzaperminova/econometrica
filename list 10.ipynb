```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.formula.api as smf
```

## 1.3

```python
df=pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
```

```python
df
```

```python
model = smf.ols(data=df, formula='output~capital+labour+wage').fit()
hat_output = model.predict()
```

```python
df['out2'] =hat_output**2
df['out3'] =hat_output**3
df['out4'] =hat_output**4
df
```

```python
model2 = smf.ols(data=df, formula='output~capital+labour+wage+out2+out3+out4').fit()
model2.params  ## сходится ли со значениями выведенными в задании
```

```python
R_un = model2.rsquared
R_res = model.rsquared
```

```python
F_obs = ((R_un - R_res)/(1-R_un))*((len(df)-7)/3) ##3 гаммы
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-7).round(2)
```

$F_{obs}>F{cr} \Rightarrow H_0 $ отвергаем
